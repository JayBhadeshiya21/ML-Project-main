2. frequency Distribution Of Categorical Columns (In Percentage %)

```python
categorical_cols = ["gender", "cholesterol", "gluc", "smoke", "alco", "active", "cardio"]

for col in categorical_cols:
    print((df[col].value_counts() / df[col].count() * 100).round(2), '\n')
```

4) Visual Analysis

Histograms (to see distribution of values)

```python
numeric_cols = ["age", "height", "weight", "ap_hi", "ap_lo"]

for col in numeric_cols:
    plt.figure(figsize=(7,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

2) Boxplots to identify Outliers

```python
numeric_cols = ['age', 'height', 'weight', 'ap_hi', 'ap_lo']
```

```python
for col in numeric_cols:
    plt.figure(figsize=(7,4))
    sns.boxplot(x=df[col])
    plt.title(f"Outliers in {col}")
    plt.show()
```

3) Frequency Distribution

```python
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    plt.title(f"Frequency Distribution of {col}")
    plt.show()
```
